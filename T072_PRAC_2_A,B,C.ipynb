***2.B Perform basic data pre-processing tasks such as handling missing values and outliers.***

```python
print("><><>< NITIN KUMAR BERA T072 ><><><")
print("2.B Perform basic data pre-processing tasks such as handling missing values and outliers.")
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer, KNNImputer

# Assuming the file '/content/test.csv' is available
df = pd.read_csv('/content/test (1).csv')
print(" Dataset Head:",df.head())
print(" Dataset Shape:", df.shape)

# Define columns with missing values for demonstration
COL_CAT = 'MSZoning'
COL_NUM_1 = 'LotFrontage'
COL_NUM_2 = 'MasVnrArea'
COL_HIGH_NAN = 'PoolQC'

print(" SECTION 1: CHECKING MISSING VALUES")
print("\nTotal missing values per column (Top 10):")
# Show only columns with missing values
missing_values = df.isnull().sum()
# Display the top 10 columns with missing values in Markdown table format
print(missing_values[missing_values > 0].sort_values(ascending=False).head(10).to_markdown(numalign="left"))
print("\nTotal number of missing values in the entire dataset:")
print(df.isnull().sum().sum())

print("\n SECTION 2: DELETING MISSING VALUES")
df_temp = df.copy()

# A. Deleting the entire row (listwise deletion)
df_dropna_row = df_temp.dropna(axis=0)
print(f"\nA. Deleting all rows with any missing value (dropna(axis=0)):")
print(f"Original rows: {df_temp.shape[0]}, Remaining rows: {df_dropna_row.shape[0]}")
print(f"Missing values in remaining data: {df_dropna_row.isnull().sum().sum()}")

# B. Deleting the entire column (for a column with many NaNs)
df_dropna_col = df_temp.drop(COL_HIGH_NAN, axis=1)
print(f"\nB. Deleting the entire column ('{COL_HIGH_NAN}'):")
print(f"Original columns: {df_temp.shape[1]}, Remaining columns: {df_dropna_col.shape[1]}")
print(f"Is '{COL_HIGH_NAN}' still present: {'PoolQC' in df_dropna_col.columns}")

print("\n SECTION 3: IMPUTING WITH PANDAS (.fillna())")
df_impute = df.copy()

# A. Replacing with an arbitrary value (for categorical 'MSZoning')
df_impute[COL_CAT] = df_impute[COL_CAT].fillna('ARBITRARY')
print(f"\nA. Imputing '{COL_CAT}' with 'ARBITRARY'. Missing values remaining: {df_impute[COL_CAT].isnull().sum()}")

# B. Replacing with the mean (for numerical 'LotFrontage')
mean_lf = df_impute[COL_NUM_1].mean()
df_impute[COL_NUM_1] = df_impute[COL_NUM_1].fillna(mean_lf)
print(f"B. Imputing '{COL_NUM_1}' with MEAN ({mean_lf:.2f}). Missing values remaining: {df_impute[COL_NUM_1].isnull().sum()}")

# C. Replacing with the median (for numerical 'MasVnrArea')
median_mv = df_impute[COL_NUM_2].median()
df_impute[COL_NUM_2] = df_impute[COL_NUM_2].fillna(median_mv)
print(f"C. Imputing '{COL_NUM_2}' with MEDIAN ({median_mv:.2f}). Missing values remaining: {df_impute[COL_NUM_2].isnull().sum()}")

# D. Forward/Backward Fill and Interpolation (using LotFrontage)
temp_series = df[COL_NUM_1].copy()
temp_series.iloc[10:15] = np.nan # Introduce a small block of NaNs for clear demo
print("\nD. Forward/Backward Fill and Interpolation (on a subset of LotFrontage):")
print(f"Forward-Fill (ffill) example (first imputed value): {temp_series.ffill().iloc[10]:.2f}")
print(f"Backward-Fill (bfill) example (first imputed value): {temp_series.bfill().iloc[10]:.2f}")
print(f"Linear Interpolation example (first imputed value): {temp_series.interpolate().iloc[10]:.2f}")

print("\n SECTION 4: IMPUTING CATEGORICAL FEATURES (Sci-kit Learn)")
# A. Impute the Most Frequent Value ('MSZoning')
imputer_mode = SimpleImputer(strategy='most_frequent')
mszoning_mode_imputed = imputer_mode.fit_transform(df[[COL_CAT]])
most_frequent_value = imputer_mode.statistics_[0]
print(f"\nA. Impute with Most Frequent value ('{COL_CAT}'):")
print(f"Mode used for imputation: '{most_frequent_value}'")
print(f"First 5 imputed values: {mszoning_mode_imputed[:5].flatten()}")

# B. Impute the Value "Missing" ('MSZoning')
imputer_const = SimpleImputer(strategy='constant', fill_value='MISSING_CAT')
mszoning_const_imputed = imputer_const.fit_transform(df[[COL_CAT]])
print("\nB. Impute with Constant value 'MISSING_CAT' (First 5):")
print(f"First 5 imputed values: {mszoning_const_imputed[:5].flatten()}")

print("\n SECTION 5: IMPUTING NUMERICAL FEATURES (Multivariate KNNImputer)")
# Impute 'LotFrontage' based on 'LotArea' and 'YearBuilt'
impute_cols = [COL_NUM_1, 'LotArea', 'YearBuilt']
X_knn = df[impute_cols].copy()

# Use KNNImputer
impute_knn = KNNImputer(n_neighbors=5)
X_knn_imputed = impute_knn.fit_transform(X_knn)

# Compare a row that was originally missing (e.g., row 10 has LotFrontage NaN)
missing_row_index = df[df[COL_NUM_1].isnull()].index[0] # First row where LotFrontage is NaN
original_lotfrontage_mean = df[COL_NUM_1].mean()
print(f"Original Row {missing_row_index} values (NaN is {COL_NUM_1}):")
print(df.iloc[missing_row_index][impute_cols].to_string())
print(f"\nImputed Row {missing_row_index} value for '{COL_NUM_1}': {X_knn_imputed[missing_row_index][0]:.2f}")
print(f"(For comparison, overall mean was: {original_lotfrontage_mean:.2f})")

print("\n SECTION 6: USING MISSINGNESS AS A FEATURE")
# Impute 'LotFrontage' using mean and add an indicator column
imputer_indicator = SimpleImputer(strategy='mean', add_indicator=True)
# The output is two columns: imputed LotFrontage and the indicator (1 if missing, 0 if present)
lotfrontage_with_indicator = imputer_indicator.fit_transform(df[[COL_NUM_1]])
print("Imputing LotFrontage with mean and adding a missing indicator column:")
print("Column 0: Imputed LotFrontage, Column 1: Missing Indicator (1=Missing)")
print(f"First 5 rows:\n{lotfrontage_with_indicator[:5]}")
print(f"\nExample row that was missing (Index {missing_row_index}):")
print(lotfrontage_with_indicator[missing_row_index])
```

## ***2 . C Manipulate and transform data using functions like filtering, sorting, and grouping.***

```python
import numpy as np
import pandas as pd

print("><><>< NITIN KUMAR BERA T072 ><><><")
print("2.C Manipulate and transform data using funcons like filtering, sorng, and grouping.")
# Assuming the file '/content/test.csv' is available
df = pd.read_csv('/content/test (1).csv')

print(" Dataset Head:",df.head())
print(" Dataset Shape:", df.shape)

# Use a clean copy for operaons that modify the DataFrame structure (like renaming, dropping)
df_working = df.copy()

print("\n--- C.2 Describing the Data (Dimensions and Stascs) ---")
print(f"Dimensions (rows, columns): {df_working.shape}")
print("\nKey numerical stascs (LotArea, OverallQual):")
print(df_working[['LotArea', 'OverallQual']].describe().to_markdown(numalign="le"))

print("\n--- C.3 Navigang DataFrames ---")
print("\nA. Select the first row (iloc[0]):")
print(df_working.iloc[0][['Id', 'MSZoning', 'LotArea']].to_string())

# Set 'Id' as index and select by index name (loc)
df_indexed = df_working.set_index('Id')
print("\nB. Select a row by index name (Id=1463) aer seng 'Id' as index (loc[1463]):")
print(df_indexed.loc[1463][['MSZoning', 'LotArea', 'YearBuilt']].to_string())

print("\n--- C.4 Selecng Rows Based on Condionals ---")
# Find houses with Excellent Overall Quality (8 or higher) and large LotArea (> 20000 sq )
conditional_df = df_working[(df_working['OverallQual'] >= 8) & (df_working['LotArea'] > 20000)]
print(f"Total matching houses: {conditional_df.shape[0]}")
print("Matching observaons (First 3):")
print(conditional_df[['OverallQual', 'LotArea', 'YearBuilt']].head(3).to_markdown(index=False, numalign="le"))

print("\n--- C.5 Replacing Values ---")
# Replace 'TA' (Typical/Average) with 'Average' in ExterQual
exter_qual_replaced = df_working['ExterQual'].replace('TA', 'Average')
print("ExterQual aer replacing 'TA' with 'Average' (First 5):")
print(exter_qual_replaced.head().to_markdown(numalign="le"))

print("\n--- C.6 Renaming Columns ---")
# Rename 'GrLivArea' to 'GroundLivingArea'
df_renamed = df_working.rename(columns={'GrLivArea': 'GroundLivingArea'})
print("Column names aer renaming (showing Id and the renamed column):")
print(df_renamed[['Id', 'GroundLivingArea']].head(2).to_markdown(index=False, numalign="le"))

print("\n--- C.7 Finding Min, Max, Mean, Count (on 'GrLivArea') and Variance/Std (on 'LotArea') ---")
print(f"Maximum GrLivArea: {df_working['GrLivArea'].max()}")
print(f"Mean GrLivArea: {df_working['GrLivArea'].mean():.2f}")
print(f"Standard Deviaon of LotArea: {df_working['LotArea'].std():.2f}")
print(f"Skewness of LotArea: {df_working['LotArea'].skew():.2f}")

print("\n--- C.8 Finding Unique Values (on 'MSZoning') ---")
print("A. Unique values (unique()):")
print(df_working['MSZoning'].unique())
print("\nB. Count of each unique value (value_counts()):")
print(df_working['MSZoning'].value_counts().to_markdown(numalign="le"))

print("\n--- C.9 Handling Missing Values (on 'LotFrontage') ---")
# Select the first 3 rows where 'LotFrontage' is missing (isnull())
missing_lotfrontage = df_working[df_working['LotFrontage'].isnull()].head(3)
print("First 3 observaons where LotFrontage is missing:")
print(missing_lotfrontage[['Id', 'MSZoning', 'LotFrontage']].to_markdown(index=False, numalign="le"))

print("\n--- C.10 Deleng a Column ('Alley') ---")
# Delete 'Alley' column (axis=1)
df_after_drop_col = df_working.drop('Alley', axis=1)
print(f"Original column count: {df_working.shape[1]}, New column count: {df_after_drop_col.shape[1]}")

print("\n--- C.11 Deleng a Row ---")
# A. Deleng the first row by index
df_after_drop_row_index = df_working[df_working.index != 0]
print(f"A. Deleng row by index (index 0): New first row Id is {df_after_drop_row_index.iloc[0]['Id']}")
# B. Deleng rows by value (remove all houses in the 'C (all)' zone)
df_after_drop_row_value = df_working[df_working['MSZoning'] != 'C (all)']
print(f"B. Deleng rows by value ('MSZoning' != 'C (all)'): {df_working.shape[0]} rows original, {df_after_drop_row_value.shape[0]} rows remaining.")

print("\n--- C.13 Grouping Rows by Values (OverallQual and GrLivArea) ---")
# Group by 'OverallQual' and find the mean of 'GrLivArea'
mean_grlivarea = df_working.groupby('OverallQual')['GrLivArea'].mean().round(0).astype(int)
print("Mean Ground Living Area (sq ) grouped by Overall Quality:")
print(mean_grlivarea.to_markdown(numalign="le"))

print("\n--- C.15 Looping Over a Column ('MSZoning') ---")
print("A. For loop (First 3 values, converted to lowercase):")
for zoning in df_working['MSZoning'].head(3):
    print(zoning.lower())
print("\nB. List comprehension (First 3 values, converted to uppercase):")
zoning_upper = [zoning.upper() for zoning in df_working['MSZoning'].head(3)]
print(zoning_upper)

print("\n--- C.16 Applying a Funcon to a Column ('YearBuilt') ---")
def age_of_house(year):
    """Calculates the age of the house relave to the last reported sales year (2010)"""
    return 2010 - year

house_age = df_working['YearBuilt'].apply(age_of_house).head(5)
print("Applied funcon to calculate house age (2010 - YearBuilt) (First 5):")
print(house_age.to_markdown(numalign="le"))

print("\n--- C.17 Applying a Funcon to Groups (Grouping by ExterQual) ---")
# Group by 'ExterQual' and calculate the size (count of rows) in each group
group_sizes = df_working.groupby('ExterQual').apply(lambda x: x.shape[0])
print("Number of houses (rows) in each ExterQual group:")
print(group_sizes.to_markdown(numalign="le"))

print("\n--- C.18 Concatenang DataFrames ---")
# Create a small dummy DataFrame represenng new data
dummy_data = {'Id': [3000, 3001], 'MSZoning': ['NEW', 'NEW'], 'LotArea': [1000, 2000]}
df_new_rows = pd.DataFrame(dummy_data)
# Concatenate the first 2 rows of the original df with the new rows (Vercally)
df_concatenated = pd.concat([df_working[['Id', 'MSZoning', 'LotArea']].head(2), df_new_rows], ignore_index=True)
print("Concatenated DataFrame (2 original rows + 2 new rows):")
print(df_concatenated.to_markdown(numalign="le"))
```
